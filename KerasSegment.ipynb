```python
def imshow(img):
    import cv2
    import IPython
    _,ret = cv2.imencode('.jpg', img) 
    i = IPython.display.Image(data=ret)
    IPython.display.display(i)
    
# image file을 보여주고자 할 때    
# from IPython.display import Image
# Image(filename='out.png')     
```

```python
def make_h_collage(images):
    import cv2
    import numpy as np

    black_space = np.zeros((200,10,3))
    white_space = np.ones((200,10,3)) *255
    
    stacked_img = np.ones(shape=(200,1,3)) * 255
    for image in images:
        img = cv2.resize(cv2.imread(image), (200,200))
        img = np.concatenate((img, white_space), axis=1)
        stacked_img = np.hstack([stacked_img, img])
        
    return stacked_img    
```

```python
def make_v_collage(img1, img2):
    import numpy as np
    h, w, c = img1.shape
    space = np.ones([10, w, c]) * 256
    stacked_img = np.vstack([img1, space, img2])
    return stacked_img
```

```python
def list_onlyfiles(directory):
    from os import listdir
    from os.path import isfile, join
    from natsort import natsorted
    
    files = [f for f in listdir(directory) if isfile(join(directory, f))]
    files = natsorted (files)
    return files
```

```python
def list_fullfilepath(directory):
    from os import listdir
    from os.path import isfile, join
    from natsort import natsorted
    files = [join(directory, f) for f in listdir(directory) if isfile(join(directory, f))]
    files = natsorted (files)
    return files
```

```python
def batch_jobs(in_path, out_path, model):
    image_files = list_onlyfiles(in_path)
    for f in image_files:
#         print(f'input: {in_path}/{f}')
#         print(f'output: {out_path}/seg_{f}.png')
        out = model.predict_segmentation(
                inp=f'{in_path}/{f}',
                out_fname=f'{out_path}/seg_{f}',
                overlay_img=False,
                show_legends=False,
                class_names = ["Sky","Building", "Pole","Road","Pavement","Tree","SignSymbol", "Fence", "Car","Pedestrian", "Bicyclist"]
                )
        
```

class_names parameter matters the order of array items. That is, the order is already fixed, and the string is just texts to be represented in the result. So, you have to find appropriate order of classes.

output 폴더가 미리 만들어져 있어야 정상적으로 파일이 생성된다.

```python
model1 = pspnet_101_voc12() # load the pretrained model trained on Pascal VOC 2012 dataset
model2 = pspnet_50_ADE_20K() # load the pretrained model trained on ADE20k dataset
model3 = pspnet_101_cityscapes() # load the pretrained model trained on Cityscapes dataset
```

```python
models = [model1, model2, model3]
models_name = ['pspnet_101_voc12','pspnet_50_ADE_20K','pspnet_101_cityscapes']
```

```python
i=0
for model in models:
    batch_jobs("./images", "./output", model)
    files=list_fullfilepath("./images")
    input_collage=make_h_collage(files)
    files=list_fullfilepath("./output")
    output_collage=make_h_collage(files)
    
    result = make_v_collage(input_collage, output_collage)
    print(models_name[i])
    imshow(result)
    i=i+1
```
